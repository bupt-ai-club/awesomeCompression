## 1.模型剪枝实践



按照剪枝粒度进行划分，剪枝分为非结构化剪枝和结构化剪枝。
- 非结构化剪枝：修剪参数的单个元素。
- 结构化剪枝：有选择性地移除网络的较大部分，比如一层或一个通道。

## 向量级别剪枝

```python
import torch 
# 创建一个矩阵weight
weight = torch.rand(4, 3)
print(f"剪枝前weight:  {weight}")
```

```python
# 将idx所在的行及列剪枝
def vector_level_pruning(weight, idx):
    row, col = idx
    prune_weight = weight.clone()
    prune_weight[row, :] = 0
    prune_weight[:, col] = 0
    return prune_weight
idx = (1, 1)
prune_weight = vector_level_pruning(weight, idx)
print(f"剪枝后weight:  {prune_weight}")
```
